```python

class BinaryTree:
    # 멤버 변수: 앞에 'self'가 붙어있는 변수. 클래스 전체에서 사용할 수 있다.
# 메서드의 self로 전달되는 것은 인스턴스 자체
##### self는 인스턴스 자기 자신

    # 생성자에서 이진 트리의 노드를 만든다.
    def __init__(self, data=None):
        self.left = None  # 왼쪽 자식 노드
        self.data = data  # 데이터
        self.right = None # 오른쪽 자식 노드
    # left와 right는 각각 루트의 자식 노드를 가리킨다


    # 트리에 노드를 삽입하는 함수
    def insert(self, data):
        print('트리에 삽입하려는 데이터 {}의 부모 노드는 {}입니다.'.format(data, self.data))
        
        ## self.data : self는 인스턴스 자기 자신이므로, root.data이므로 12이다.
        
        # 데이터가 부모 노드의 데이터보다 작으면 왼쪽에 크면 오른쪽에 삽입한다.
        if data < self.data:
            # 왼쪽 자식 노드(self.left)가 비어있어야 삽입 ㄱㄴ.
            if self.left is None:
                print('부모({}) 왼쪽에 데이터 {} 추가 가능'.format(self.data, data))
                
            ## 왼쪽 자식 링크(self.left)에, 데이터 data를 가지는 새로운 이진 트리 노드를 생성하여 연결합니다.(ㄹㄹ)
            ## self.left : 부모의 왼쪽 노드 전체를 의미한다(left,data,right 다 포함;)
                self.left = BinaryTree(data)
                print('부모({}) 왼쪽에 데이터 {} 추가 완료'.format(self.data, data))
            else:
                # 부모 노드의 왼쪽 링크가 비어있지 않기 때문에 데이터를 추가할 수 없다.
                print('부모({}) 왼쪽에 {} 추가가 불가능하므로 부모 왼쪽 노드 {}에서 insert() 함수를 실행한다.'.format(
                    self.data, data, self.left.data 
                ))
                self.left.insert(data)
                
        elif data > self.data:
            print(' data > self.data -> 부모 노드의 오른쪽에 삽입한다.')
            # 부모 노드의 오른쪽 링크(self.right)가 비어(None)있어야 데이터를 삽입할 수 있다.
            if self.right is None:
                print('부모({}) 오른쪽에 데이터 {} 추가 가능'.format(self.data, data))
                
                ## 오른쪽 자식 링크에, 새로운 이진 트리 노드를 생성하여 연결합니다.(ㄹㄹ)
                self.right = BinaryTree(data)
                print('부모({}) 오른쪽에 데이터 {} 추가 완료'.format(self.data, data))
            else:
                print('부모({}) 오른쪽에 {} 추가가 불가능하므로 부모 오른쪽 노드 {}에서 insert() 함수를 실행한다.'.format(
                    self.data, data, self.right.data
                ))
                self.right.insert(data)
        
        # 트리에 삽입하려는 데이터가 트리에 존재하는가?
        else:
            print('{}는(은) 트리에 존재하는 데이터입니다.'.format(data))
    # ===== insert
    
'''
이진 트리의 운행(같은 노드를 중복해서 방문하지 않고 모든 노드를 한 번씩 방문하는 방법)

# preorder  : root => left => right: 12 => 6 => 3 => 10 => 20 => 15 => 25
    루트를 먼저 확인한 후에 자식 노드를 확인하는 순회 방법이다.
  
# inorder   : left => root => right: 3 => 6 => 10 => 12 => 15 => 20 => 25: 오름차순 정렬
    중위순회는 왼쪽 자식 노드, 루트 노드, 오른쪽 자식 노드 순으로 값을 확인하는 방식이다.

# postorder : left => right => root: 3 => 10 => 6 => 15 => 25 => 20 => 12
    후위 순회는 자식 노드를 모두 확인한 후에 루트 노드를 확인하는 순회 방법이다.


'''    

    # preorder : 루트를 먼저 확인한 후에 자식 노드를 확인하는 순회 방법이다.
    # root => left => right : 12 => 6 => 3 => 10 => 20 => 15 => 25
    def preorder(self):
        print('{}의 preorder() 함수 시작'.format(self.data))
        print('방문한 노드 {}'.format(self.data))
        # 왼쪽 자식 노드가 있나 검사해서 있으면 자식 노드에서 preorder() 함수를 재귀적으로 실행한다.
        if self.left:
            self.left.preorder()
        # 오른쪽 자식 노드가 있나 검사해서 있으면 자식 노드에서 preorder() 함수를 실행한다.
        if self.right:
            self.right.preorder()
        print('{}의 preorder() 함수 끝'.format(self.data))
    # ===== preorder        
    
    # inorder : 왼쪽 자식 노드, 루트 노드, 오른쪽 자식 노드 순으로 값을 확인하는 방식이다.
    # left => root => right: 3 => 6 => 10 => 12 => 15 => 20 => 25: 오름차순 정렬    
    def inorder(self):
        print('{}의 inorder() 함수 시작'.format(self.data))
        if self.left:
            self.left.inorder()
        print('방문한 노드 {}'.format(self.data))
        if self.right:
            self.right.inorder()
        print('{}의 inorder() 함수 끝'.format(self.data))
        
    # postorder : 루트 노드를 마지막에 확인한다.
    # left => right => root: 3 => 10 => 6 => 15 => 25 => 20 => 12
    def postorder(self):
        print('{}의 postorder() 함수 시작'.format(self.data))
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        
        print('방문한 노드 {}'.format(self.data))
        print('{}의 postorder() 함수 끝'.format(self.data))

```

```python
# 이진 트리의 최상위(root) 노드를 만든다.
root = BinaryTree(12) # 객체 생성
# 이진 트리에 데이터(노드)를 추가한다.
root.insert(12)
print('=' * 80)
root.insert(6)
print('=' * 80)
root.insert(20)
print('=' * 80)

root.insert(3)
print('=' * 80)
root.insert(10)
print('=' * 80)
root.insert(15)
print('=' * 80)
root.insert(25)
print('=' * 80)
# root.insert(55)
```
