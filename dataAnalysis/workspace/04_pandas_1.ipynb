데이터 분석을 위해서 데이터를 불러온다. => 불러온 데이터를 데이터 집합(set)이라고 한다.

gapminder : 무료 데이터 가져오기 좋은 사이트

```python
# read_csv() : 데이터 집합을 불러와서 데이터프레임으로 리턴한다.
# read_csv() :  기본값 sep = ','로 구분된 데이터를 불러온다.
gapminder_df = pd.read_csv('./data/gapminder.tsv', sep='\t') # 구분자=탭
print(type(gapminder_df)) # <.DataFrame'>
gapminder_df
```

```python
# head() 함수는 데이터프레임 앞쪽 데이터를 지정한 인수 개만큼 출력한다. 기본값은 5
gapminder_df.head(3)
```

```python
# tail() 함수는 데이터프레임 뒤쪽 데이터를 지정한 인수 개만큼 출력한다. 기본값은 5
gapminder_df.tail(3)
```

```python
# shape 데이터프레임의 행과 열 정보.
print(gapminder_df.shape)
print(gapminder_df.shape[0]) #  행의 개수
print(gapminder_df.shape[1]) #  열의 개수

# columns  데이터프레임의 열 이름을 저장하고 있다.
print(gapminder_df.columns)
# country: 국가, continent: 대륙, year: 년도, lifeExp: 평균수명, pop: 인구수, gdpPercap: 1인당 gdp

# dtypes : 데이터의 자료형. object => 문자열
print(gapminder_df.dtypes)
```

```python
# info(): dtypes 보다 좀 더 자세한 내용을 확인할 수 있다.
print(gapminder_df.info())
```

열 단위 데이터 추출하기


```python
# 1개의 열만 추출하면 시리즈를 얻을 수 있다.

#1, 2 : 딕셔너리에서 key로 value 가져오는것과 같음 
# country_df = gapminder_df['country'] #1번째.

# 2. get() 메소드의 인수로 열 이름을 지정.
# country_df = gapminder_df.get('country')

country_df = gapminder_df.country #  데이터프레임 이름 뒤에 '.'을 찍고 열 이름을 적어주면 된다.
print(type(country_df)) # <Series'>
print(country_df.head(3))
print(country_df.tail(3))
```

```python
# 2개 이상의 열을 추출하면 데이터프레임을 얻을 수 있다.

# 2개 이상의 열을 얻어오려면 [] 안에 반드시 리스트 형태로 열 이름을 적어줘야 한다.
subset_df = gapminder_df[['country', 'continent', 'year']]
# subset_df = gapminder_df.get(['country', 'continent', 'year'])
# subset_df = gapminder_df.['country', 'continent', 'year'] # 에러, 1개의 열만 얻어올 때 사용해야 한다.
print(type(subset_df)) # <DataFrame'>
subset_df
 
```
