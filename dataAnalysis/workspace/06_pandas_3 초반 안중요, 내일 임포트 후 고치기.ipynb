concat() 함수는 하나의 데이터프레임에 한 번에 2개 이상이 데이터프레임을 연결하는 함수이다. 만약에 데이터프레임에 연결할 데이터프레임이 1개라면 append() 메소드를 사용해도 된다.

```python
row_concat = pd.concat([df1, df2, df3], axis=0, ignore_index=True)
row_concat
```

 append() 메소드와 딕셔너리를 사용하면 추가할 데이터를 데이터프레임으로 구성하지 않고도 데이터프레임에 추가할 수 있다.
데이터프레임.append(추가할 데이터가 저장된 딕셔너리)

```python
data_dict = {
    'A': 'n5',
    'B': 'n6',
    'C': 'n7',
    'D': 'n8'
}
print(data_dict)
# df1.append(data_dict) # 에러
# 데이터프레임이 아닌 딕셔너리를 추가하기 때문에 딕셔너리에는 인덱스가 없으므로 ignore_index=True 옵션을 지정해서
# 기존 인덱스를 무시하고 다시 부여하게 해야 정상적으로 딕셔너리의 데이터가 데이터프레임에 추가된다.
df1.append(data_dict, ignore_index=True)
```

데이터프레임을 오른쪽으로 연결하려면 concat() 메소드의 axis 속성을 1로 지정하면 된다.
axis 속성을 생략하면 axis 속성의 기본값인 0이 사용되면 아래쪽으로 추가되고 1을 사용하면 오른쪽으로 추가된다.

```python
col_concat = pd.concat([df1, df2, df3], axis=1, ignore_index=True)
col_concat
```

```python
col_concat = pd.concat([df1, df2, df3], axis=1)
col_concat
```

```python
# 데이터프레임['열이름'] = [데이터] 형태로 실행하면 새로운 열을 추가 또는 수정할 수 있다.
col_concat['new_col'] = ['b1', 'b2', 'b3', 'b4'] # 없는 열에 데이터를 넣으면 추가된다.
col_concat
```

데이터프레임의 열 이름을 변경하려면 columns 속성에 새 열 이름을 리스트 형태로 전달하면 된다.



```python
print(df1.columns)
df2.columns = ['E', 'F', 'G', 'H']
print(df2.columns)
df3.columns = ['A', 'C', 'F', 'H']
print(df3.columns)
```

```python
# 새롭게 데이터프레임의 열 이름을 부여하고 concat() 메소드로 연결하면 연결되는 데이터프레임에 없는 열 이름의
# 데이터는 누락값으로 처리된다.
row_concat = pd.concat([df1, df2, df3], ignore_index=True)
row_concat
```

concat() 메소드의 join 속성을 inner로 지정하면 연결되는 데이터프레임에 공통으로 데이터가 있는 열만 골라서 연결할 수 있고 이 경우에는 누락값이 발생되지 않는다.



```python
# join 속성의 기본값은 outer이고 공통으로 데이터가 있는 열에 신경쓰지 않고 모두 연결한다. => NaN이 대량 발생된다.
row_concat = pd.concat([df1, df2, df3], ignore_index=True, join='outer')
row_concat
```

```python
# df1, df2, df3 데이터프레임은 3개 모두 공통 열이 없기 때문에 빈 데이터프레임이 생성된다.
row_concat = pd.concat([df1, df2, df3], ignore_index=True, join='inner')
row_concat
```

```python
# df1, df3 데이터프레임은 2개 모두 공통 열이 있기 때문에 공통 열을 연결한 데이터프레임이 생성된다. => A, C열이 연결된다.
row_concat = pd.concat([df1, df3], ignore_index=True, join='inner')
row_concat
```

```python
# df2, df3 데이터프레임은 2개 모두 공통 열이 있기 때문에 공통 열을 연결한 데이터프레임이 생성된다. => F, H열이 연결된다.
row_concat = pd.concat([df2, df3], ignore_index=True, join='inner')
row_concat
```

```python
# 데이터프레임의 인덱스를 변경하려면 index 속성에 새 인덱스를 리스트 형태로 지정하면 된다.

print(list(df1.index))
df2.index = [4, 5, 6, 7]
print(list(df2.index))
df3.index = [0, 2, 5, 7]
print(list(df3.index))
```

```python
row_concat = pd.concat([df1, df2, df3], axis=1, ignore_index=True, join='outer')
row_concat
```
