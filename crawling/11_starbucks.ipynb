```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
import json
```

```python
# 시도코드 크롤링

targetSite = 'https://www.starbucks.co.kr/store/getSidoList.do'

# post() 메소드 : data속성은 전송할 데이터가 없으면 생략ㄱㄴ
request = requests.post(targetSite, data={ # 넘어가는 data: Payload에 있음.
    'rndCod': 'W0VMMU0TEP'  # 얜 전달 안해도 돼.
})
# print(request)

starbucks = request.text
# print(starbucks)
# print(type(starbucks)) # <class 'str'> -> (리스트 또는 딕셔너리)로 변환해서 처리한다
# json.loads(starbucks) or request.json() 해주기

# sidoList = json.loads(starbucks)
sidoList= request.json() #  딕셔너리로 변환
# print(type(sidoList)) # <class 'dict'>
# print(sidoList)


# sidoList에는 문자열 형태의 json 데이터가 파이썬이 처리할 수 있는 딕셔너리 형태로 변경돼 저장되었다.
# 딕셔너리로 변환된 데이터에서 'list'라는 key에 할당된 시도 정보를 얻어온다. 
# print(sidoList.get('list'))
# print(type(sidoList.get('list'))) # <class 'list'>
# print(sidoList['list']) # 리스트 키 입력해서 값 꺼내기 
# print(type(sidoList['list'])) # <class 'list'>


# 시도코드의 개수만큼 반복하며 
# key : 시도코드('sido_cd'), value : 시도이름('sido_nm')로 하는 딕셔너리를 만든다.
sido_list = {}
for sido in sidoList['list']:
#     print(type(sido)) #<class 'dict'>
#     print('{}: {}'.format(sido.get('sido_cd'), sido['sido_nm']))
    # 딕셔너리이름[key] = value
    sido_list[sido.get('sido_cd')] = sido['sido_nm']
print(sido_list)
```

```python
# 구군코드 크롤링

sido_cd = input('구군코드를 검색하려는 시도코드를 입력하세요: ')
targetSite = 'https://www.starbucks.co.kr/store/getGugunList.do'
request = requests.post(targetSite, data={
    'sido_cd': sido_cd 
})
# print(request)

# print(type(request.text)) #<class 'str'>
# print(type(request.json())) #<class 'dict'>로 변환.

gugunList = request.json()
# 데이터 테스트~ 
# print(len(gugunList))
# print(gugunList.keys()) # dict_keys(['list'])
# print(len(gugunList['list'])) # 시별 구의 개수

# 구군코드의 개수만큼 반복하며 구군코드('gugun_cd')를 key로 하고 구군이름('gugun_nm')을 value로 하는 딕셔너리를 만든다.
gugun_list = {}
for gugun in gugunList['list']:
    gugun_list[gugun.get('gugun_cd')] = gugun['gugun_nm']
print(gugun_list)
```
