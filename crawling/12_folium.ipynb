```python
'''
folium은 python에서 제공하는 지도를 다루는 대표적인 라이브러리 입니다. 
지도를 생성하고 Marker를 추가하여 시각화하거나 
원으로 범위를 표기하고 html 파일로 내보내기 등을 수행할 수 있습니다
'''
# !pip install folium
import folium
```

```python
# folium.Map(location=[위도, 경도], zoom_start=숫자)
foliumMap = folium.Map(location=[37.5662952, 126.9779451], zoom_start=18)

# folium 라이브러리의 Marker() 메소드의 location 속성으로 지정한 위치에 popup 속성으로 지정한 문자열을 표시하는 marker를
# 만든다.
# 작성된 marker는 add_to() 메소드를 사용해서 folium 라이브러리로 생성한 지도에 추가한다.
# folium.Marker(location=[37.5662952, 126.9779451]).add_to(foliumMap)
# folium.Marker(location=[37.5662952, 126.9779451], popup='서울특별시청').add_to(foliumMap)

# marker를 클릭했을 때 팝업되는 말풍선의 크기가 마음에 들지 않으면 folium 라이브러리의 Popup() 메소드에서 max_width
# 속성으로 말풍선의 최대 가로 크기를 지정할 수 있다.
popup = folium.Popup('서어어어어어어울 시이이이이이이처어어어어어어어엉', max_width=300)
folium.Marker(location=[37.5662952, 126.9779451], popup=popup).add_to(foliumMap)

# jupyter notebook에서 popup으로 표시되는 한글이 깨져서 보일 수 있다.

# save() 메소드를 통해 html 형식으로 저장이 가능합니다. 
# folium.Map([ ... ]).save('파일이름.html')
foliumMap.save('./foliumMap.html')

foliumMap
```
